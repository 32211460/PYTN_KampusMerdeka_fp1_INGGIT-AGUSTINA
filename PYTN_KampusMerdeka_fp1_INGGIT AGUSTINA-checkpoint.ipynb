# Implementasi Linear Regression dengan Scikit-Learn

Import the required libraries

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

Load data

```python
df = pd.read_csv('data uber vs lyft.csv')
```

Prepare data

```python
X = df[['distance', 'surge_multiplier', 'name']]
X = pd.get_dummies(X, columns=['name'], prefix=['name'])
y = df['price']
```

Handle NaN values

```python
X = X.fillna(X.mean())
y = y.fillna(y.mean())
```

Split data into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

Train model

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

Predict

```python
y_pred = lr_model.predict(X_test)
```

Evaluate model

```python
score = lr_model.score(X_test, y_test)
print(f"R-squared score: {score}")
```

# Analysis

Deskripsi Model:
Model di atas merupakan model Linear Regression yang digunakan untuk memprediksi harga menggunakan dataset "data uber vs lyft.csv". Model Linear Regression adalah model statistik yang digunakan untuk memodelkan hubungan linier antara variabel independen (fitur) dan variabel dependen (target). Tujuannya adalah untuk memprediksi nilai target berdasarkan nilai fitur-fitur yang diberikan. Model Linear Regression digunakan dalam berbagai bidang, termasuk ekonomi, ilmu sosial, keuangan, dan machine learning.

Informasi model Linear Regression:

Fitur yang digunakan: Fitur yang digunakan dalam model ini adalah distance, surge_multiplier, dan name. Fitur distance dan surge_multiplier merupakan fitur numerik, sedangkan name merupakan fitur kategorikal yang diubah menjadi variabel dummy menggunakan one-hot encoding. 

Preprocessing: Data yang hilang (missing values) dalam fitur-fitur X diisi dengan nilai rata-rata menggunakan fillna(X.mean()), sedangkan nilai target y diisi dengan nilai rata-rata menggunakan fillna(y.mean()). Ini memastikan bahwa model tidak terganggu oleh adanya data yang hilang.

Pembagian Dataset: Dataset dibagi menjadi data latih (X_train dan y_train) dan data uji (X_test dan y_test) dengan rasio 80:20 menggunakan train_test_split. Data uji digunakan untuk mengevaluasi performa model yang telah dilatih.

Pelatihan Model: Model Linear Regression (LinearRegression()) dilatih dengan menggunakan data latih (X_train dan y_train) menggunakan fit(X_train, y_train). Model belajar untuk menyesuaikan pola antara fitur-fitur dan target harga.

Evaluasi Model: Performa model dievaluasi menggunakan R-squared score. R-squared score mengukur sejauh mana variasi dalam target harga dapat dijelaskan oleh fitur-fitur yang ada dalam model. Semakin tinggi nilai R-squared score, semakin baik model dalam menjelaskan variasi data. R-squared score dihitung menggunakan score(X_test, y_test).

Kesimpulan model linear regression:

R-squared score yang tinggi (0.9176781468354887) menunjukkan bahwa model yang digunakan dalam analisis regresi linear dengan dataset "uber vs lyft.csv" cukup baik dalam menjelaskan variabilitas dalam data. R-squared score mengukur seberapa baik variabel independen (X) dapat menjelaskan variasi dalam variabel dependen (y). Semakin mendekati nilai 1, semakin baik model tersebut menjelaskan variasi dalam data.

Dalam konteks ini, kesimpulan yang dapat ditarik adalah bahwa variabel independen (mungkin fitur-fitur seperti waktu, lokasi, atau faktor lainnya yang ada dalam dataset) memiliki hubungan yang kuat dengan variabel dependen (misalnya, perbedaan antara jumlah pengguna Uber dan Lyft). Model regresi linear dengan R-squared score sebesar 0.9176781468354887 dapat menjelaskan sekitar 91,77% variabilitas dalam data, yang menunjukkan adanya hubungan yang signifikan antara fitur-fitur yang ada dalam dataset dengan perbedaan jumlah pengguna Uber dan Lyft.
